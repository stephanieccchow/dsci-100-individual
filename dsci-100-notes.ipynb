```python
# Import packages
import altair as alt
import pandas as pd

# Simplify working with large datasets in Altair
alt.data_transformers.enable('vegafusion')
```

```python
# Import players.csv dataframe
url_players = "https://drive.google.com/uc?id=1Mw9vW0hjTJwRWx0bDXiSpYsO3gKogaPz"
players = pd.read_csv(url_players)

players
```

```python
# Import sessions.csv dataframe
url_sessions = "https://drive.google.com/uc?id=14O91N5OlVkvdGxXNJUj5jIsV5RexhzbB"
sessions = pd.read_csv(url_sessions)

sessions
```

```python
# Are the names unique?
count = pd.DataFrame(players["name"].value_counts()).reset_index()

count["count"].unique()
```

```python
# Turn start and end time into datetime (instead of object)
sessions["end_time"] = pd.to_datetime(sessions["end_time"], dayfirst=True)
sessions["start_time"] = pd.to_datetime(sessions["start_time"], dayfirst=True)

# Create column of minutes of play
sessions.assign(session_time_min = sessions["end_time"]-sessions["start_time"])
#sessions.groupby("hashedEmail").()
```

```python
players_scatterplot = (
    alt.Chart(players, title="Hours Contributed by Each Player with Age").mark_circle(opacity=0.40).encode(
        x = alt.X("age").title("Age (years)"),
        y = alt.Y("played_hours").title("Player's Total Time Played (hours)")
    )
    .configure_axis(titleFontSize=12)
)

players_scatterplot
```

```python
players_scatterplot = (
    alt.Chart(players, title="Hours Contributed by Each Player with Age and Experience").mark_point().encode(
        x = alt.X("age").title("Age (years)"),
        y = alt.Y("played_hours").title("Player's Total Time Played (hours)"),
        color = alt.Color("experience")
    )
    .configure_axis(titleFontSize=12)
)

players_scatterplot
```

```python
### TEST

#Mean total time played over age

players_mean = (
    players.groupby("age").mean("played_hours").reset_index()
)

players_scatterplot = (
    alt.Chart(players_mean, title="Title").mark_line().encode(
        x = alt.X("age").title("Age (years)"), #.scale(type="log")
        y = alt.Y("played_hours").title("Player's Total Time Played (hours)")
    )
    .configure_axis(titleFontSize=12)
)
players_scatterplot
```
